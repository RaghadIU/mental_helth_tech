Import Libraries

```python
# Import required libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
sns.set(style="whitegrid")
print("Libraries imported successfully.") # to be sure that all libraries are imported 
```

Load Dataset

```python
# Load the Original Dataset
data_path = "../data/raw/mental-heath-in-tech-2016_20161114.csv"
df = pd.read_csv(data_path)

print("Dataset loaded successfully")
print("Number of rows and columns:", df.shape)
```

 Age Distribution Plot

```python
# Example Visualization: Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['What is your age?'], bins=30, kde=True)
plt.title("Age Distribution Among Participants")
plt.xlabel("Age")
plt.ylabel("Number of Participants")
plt.show()
```
